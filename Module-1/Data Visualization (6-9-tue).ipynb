```python


# Implementation of matplotlib.pyplot.annotate()
# function
  
import matplotlib.pyplot as plt
import numpy as np
fig,ax=plt.subplots()

x=np.linspace(0,20,1000)
print(x)

ax.plot(x,np.cos(x))
ax.axis('equal')

ax.annotate('local Maximum', xy=(6.28,1), xytext=(10,4),arrowprops=dict(facecolor='green',shrink=0.04))

ax.annotate('local Minimum', xy=(5*np.pi,-1), xytext=(2,-6),arrowprops=dict(arrowstyle="->",connectionstyle="angle3,angleA=0,angleB=-90"))
```

# Adding Annotate and legend

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# you cal also write
#from matplotlib import pyplot as plt

x = np.arange(0,10)
y = x^2
z = x^3
plt.plot(x,y)

plt.plot(y,z)

plt.plot(x,z)
plt.xlabel("Time")
plt.ylabel("Distance")
plt.title("Annotation and legend- Graph Drwaing")
plt.show()
```

```python
# you cal also write
#from matplotlib import pyplot as plt

x = np.arange(0,10)
y = x^2
z = x^3
plt.plot(x,y)

plt.plot(y,z)
plt.plot(x,z)

plt.xlabel("Time")
plt.ylabel("Distance")
plt.title("Annotation and legend- Graph Drwaing")


#Annotation
plt.annotate(xy=[2,1], text="First Entry")
plt.annotate(xy=[5,5], text="Second Entry")
plt.annotate(xy=[8,9], text="Third Entry")

#add legend
plt.legend(["Race 1","Race 2", "Race 3"],loc=2)

plt.show()
```

```python
# you cal also write
#from matplotlib import pyplot as plt

x = np.arange(0,10)
y = x^2
z = x^3
plt.plot(x,y)

plt.plot(y,z)
plt.plot(x,z)

plt.xlabel("Time")
plt.ylabel("Distance")
plt.title("Annotation and legend- Graph Drwaing")


#Annotation
plt.annotate(xy=[2,1], text="First Entry", rotation=20)
plt.annotate(xy=[5,5], text="Second Entry",rotation=20)
plt.annotate(xy=[8,9], text="Third Entry", rotation=25)

#add legend
plt.legend(["Race 1","Race 2", "Race 3"],loc=4)

plt.show()
```

# pie Chart

pie chart is a circular statical plot that can display only one series of data.
the area of chart is the total % of the new given data

```python
#import libraries 
import numpy as np
import matplotlib.pyplot as plt
#creating dataset

Branch=["Computer","Electrical","Mechanical","Civil","Automobile"]
data=[50,34,40,29,12]


#create pie Chart

fig= plt.figure(figsize=(10,7))
plt.pie(data,labels=Branch)

plt.show()
```

```python
#import libraries 
import numpy as np
import matplotlib.pyplot as plt
#creating dataset

Branch=["Computer","Electrical","Mechanical","Civil","Automobile"]
data=[50,34,40,29,12]
colors=['c','fuchsia','w','darkviolet','lawngreen']
explode=[0.1,0,0,0.3,0]
#create pie Chart

fig= plt.figure(figsize=(10,7))
plt.pie(data,labels=Branch,explode=explode,autopct='%1.1f%%',colors=colors,shadow=True)

plt.show()
```

# Saving the Chart

```python
import matplotlib.pyplot as plt
import random as random
  
students = ["Twisha","Palak","Kajal","Jhanvi",
            "Aayushi","Shivam","Divyesh","Dhruv",
            "Trushan","krupa","apurva"]
  
marks=[]
for i in range(0,len(students)):
     marks.append(random.randint(0, 200))
  
plt.xlabel("Students")
plt.xticks(rotation ="vertical")
plt.ylabel("Marks")
plt.title("CLASS RECORDS")
plt.plot(students,marks,color = 'green',
         linestyle = 'solid', marker = 'o',
         markerfacecolor = 'red', markersize = 12)
plt.savefig("class.pdf",format="pdf")
plt.show()
```
